## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
# IMPORTANT NOTE:
# The functions below have been modified to add functionality, and in the course of that process
# it was realized that passing arguments between the functions is very inefficient and makes
# tuning the process harder than it needs to be. For this reason, I will ultimately construct
# a lane finding class object below, which is much cleaner and easier to use and maintain,
# since the hyperparameters are attached to the object and can therefore be used by all the
# functions without needing to be explicitly passed, and can easily be adjusted by the user.
# I have left the functions below in place only to show my process of designing the pipeline.
# For actually deploying the pipeline, I will use the class object.
````

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def get_slope_intercept(lines):
    """
    Takes in a list of lines and returns corresponding slopes and intercepts for each
    """
    slopes = []
    intercepts = []
    for line in lines:
        for x1, y1, x2, y2 in line:
            slope = float(y2 - y1) / float(x2 - x1)
            intercept = float(y2) - float(slope * x2)
            slopes.append(slope)
            intercepts.append(intercept)
    
    return np.array(slopes), np.array(intercepts)

def draw_lines(img, 
               lines, 
               color=[255, 0, 0], 
               thickness=2, 
               filter_and_merge=True,
               min_slope_thresh=0.1, 
               slope_similarity_thresh=0.1,
               intercept_similarity_thresh=40
              ):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    if filter_and_merge == True:
        lines = filter_and_merge_lines(lines,
                                 min_slope_thresh=min_slope_thresh,
                                 slope_similarity_thresh=slope_similarity_thresh,
                                 intercept_similarity_thresh=intercept_similarity_thresh
                                )
    
    for line in lines:
        #print('line shape:', line.shape)
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)
            
def filter_and_merge_lines(lines, 
                           min_slope_thresh=0.2, 
                           slope_similarity_thresh=0.1, 
                           intercept_similarity_thresh=40):
    """
    Filters out horizontal lines according to a minimum absolute slope and merges the 
    remaining lines by averaging groups formed through slope and intercept similarity
    """
    # Get slopes and intercepts
    slopes, intercepts = get_slope_intercept(lines)
    #slopes = np.array(slopes)
    #intercepts = np.array(intercepts)
    #print('slopes:', slopes)
    # Filter out horizontal lines
    filtered_indices = np.where(np.abs(slopes) >= min_slope_thresh)
    lines = np.array(lines)[filtered_indices]#.copy()
    #lines = lines[filtered_indices]
    slopes = slopes[filtered_indices]
    intercepts = intercepts[filtered_indices]
    
    # Establish an empty list to store indices of lines we have already merged
    deleted_indices = []
    # Establish an array to stack merged lines onto
    merged_lines = np.zeros((0, 1, 4), dtype='int32')
    
    # Iterate through remaining arrays and either average groups or add unique lines
    for i, (slope, intercept, line) in enumerate(zip(slopes, intercepts, lines)):
        #print('len slopes:', len(slopes))
        # Check to see if index has already been assigned to a group and skip if so
        if i in deleted_indices:
            continue
        # Break if we've merged all lines
        if len(slopes) <= 0:
            break
        #print('new slopes:', slopes)
        #print('current slope:', slope)
        #print('abs(new slopes - current slope):', np.abs(slopes-slope))
        # Get indices of lines similar in slope to current index
        slope_group_indices = np.where(np.abs(slopes - slope) < slope_similarity_thresh)[0]
        #print('slope_group_indices:', slope_group_indices)
        # Get indices of lines with similar intercepts
        #print('abs(intercepts - current intercept):', np.abs(intercepts - intercept))
        int_group_indices = np.where(np.abs(intercepts - intercept) < intercept_similarity_thresh)[0]
        # Find intersection of these groups
        group_indices = [idx for idx in slope_group_indices if idx in int_group_indices]
        #print('group_indices:', group_indices)
        # If there are no similar lines, stack current unique line and move on
        if len(group_indices) == 0:
            merged_lines = np.vstack([merged_lines, line.reshape((1, 1, 4))])
            slopes = np.delete(slopes, 0, 0)
            intercepts = np.delete(intercepts, 0, 0)
            lines = np.delete(lines, 0, 0)
            deleted_indices.append(i)
            continue
        # Otherwise average the lines in the group, stack merged line,
        # and update deleted indices accordingly
        merged_line = lines[group_indices].mean(axis=0)
        #print('merged_line:', merged_line)
        #print('merged_line shape:', merged_line.shape)
        merged_line = merged_line.round(0).astype('int32')
        #print('merged_line after round:', merged_line)
        slopes = np.delete(slopes, group_indices, 0)
        intercepts = np.delete(intercepts, group_indices, 0)
        lines = np.delete(lines, group_indices, 0)
        [deleted_indices.append(idx+i) for idx in group_indices]
        merged_lines = np.vstack([merged_lines, merged_line.reshape((1, 1, 4))])
        
    #print('merged_lines:', merged_lines)
    return merged_lines

def hough_lines(img, rho, theta, threshold, 
                min_line_len, 
                max_line_gap, 
                filter_and_merge=True, 
                min_slope_thresh=0.1, 
                slope_similarity_thresh=0.2, 
                intercept_similarity_thresh=40,
                color=[255,0,0],
                thickness=2,
                line_img=None
               ):
    """
    `img` should be the output of a Canny transform.
    
    Optional params:
    filter_and_merge - (bool) whether to filter and merge lines before drawing
    min_slope - (float) minimum slope of lines to be drawn
    slope_similarity_thresh - (float) max distance between slope values to merge lines
    intercept_similarity_thresh - (int) max number of pixels between intercepts to merge lines
        
    Returns an image with hough lines drawn.
    """
    #plt.imshow(img)
    lines = cv2.HoughLinesP(img, 
                            rho, 
                            theta, 
                            threshold, 
                            np.array([]), 
                            minLineLength=min_line_len, 
                            maxLineGap=max_line_gap,
                           )
    #print('shape of hough lines return:', lines.shape)
    #print('dtype of hough lines return:', lines.dtype)
    #return lines
    #if filter_and_merge:
    #    line_img = img.copy().astype(np.uint8)
    #else:
    #    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    if line_img is None:
        line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, 
               lines, 
               color=color, 
               thickness=thickness,
               filter_and_merge=filter_and_merge,
               min_slope_thresh=min_slope_thresh, 
               slope_similarity_thresh=slope_similarity_thresh,
               intercept_similarity_thresh=intercept_similarity_thresh
              )
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# Iterate through test_images file names, run each image through pipeline, and then save
# to directory.
def lane_finder(initial_img, 
                canny_thresh_lwr=50, 
                canny_thresh_upr=150, 
                blur_kernel_size=5,
                roi_vertices=None,
                hough_rho=1,
                hough_theta=np.pi/180,
                hough_thresh=50,
                min_line_len=50,
                max_line_gap=100,
                filter_and_merge=False,
                min_slope_thresh=0.4,
                slope_similarity_thresh=0.2,
                intercept_similarity_thresh=100
               ):
    # Generate default roi vertices if none provided
    if roi_vertices is None:
        imshape = initial_img.shape
        roi_vertices = np.array([[(0 + imshape[1] // 10, imshape[0]),
                          (imshape[1] // 2 - imshape[1] // 20, 
                           imshape[0] // 2 + imshape[0] // 10), 
                          (imshape[1] // 2 + imshape[1] // 20, 
                           imshape[0] // 2 + imshape[0] // 10),
                          (imshape[1] - imshape[1] // 10, imshape[0])]], dtype=np.int32)
    # Make grayscale copy of input image
    gray = grayscale(initial_img.copy())
    # Blur image
    blur_gray = gaussian_blur(gray, blur_kernel_size)
    # Get Canny edge detection
    edges = canny(blur_gray, canny_thresh_lwr, canny_thresh_upr)
    # Make mask and get RoI
    masked_edges = region_of_interest(edges, roi_vertices)
    # Run masked edges through Hough transform to get lines
    if filter_and_merge:
        line_img = initial_img
        color = [255,0,255]
        thickness = 4
    else:
        line_img = None
        color = [255,0,0]
        thickness = 2
    line_img = hough_lines(masked_edges, 
                           hough_rho,
                           hough_theta,
                           hough_thresh,
                           min_line_len,
                           max_line_gap,
                           filter_and_merge=filter_and_merge,
                           min_slope_thresh=min_slope_thresh,
                           slope_similarity_thresh=slope_similarity_thresh,
                           intercept_similarity_thresh=intercept_similarity_thresh,
                           color=color,
                           thickness=thickness,
                           line_img=line_img
                          )
    #return line_img
    if filter_and_merge:
        return line_img
    else:
        return weighted_img(line_img, initial_img)
    
````

````python
# This pipeline is working, but due to the structure of the preconstructed functions, the
# passing of parameters through these functions has become unweildy, and would be much better
# facilitated by a class object which can store these options as hyperparameters. Rather than
# trying to work the extrapolation functionality into the above pipeline, I will just construct
# this class object now, since it will be much cleaner and more maintainable.

class Lane_Finder:
    
    def __init__(self,
                 canny_thresh_lwr=50,
                 canny_thresh_upr=150,
                 blur_kernel_size=5,
                 roi_vertices=None,
                 hough_rho=1,
                 hough_theta=np.pi/180,
                 hough_thresh=50,
                 min_line_len=50,
                 max_line_gap=100,
                 filter_and_merge=False,
                 min_slope_thresh=0.4,
                 slope_similarity_thresh=0.2,
                 intercept_similarity_thresh=100,
                 extrapolate_lines=False,
                 lane_horizon=0.60,
                 side_margin=0.05,
                 line_color=[255,0,255],
                 line_thickness=4
                ):
        self.canny_thresh_lwr = canny_thresh_lwr
        self.canny_thresh_upr = canny_thresh_upr
        self.blur_kernel_size = blur_kernel_size
        self.roi_vertices = roi_vertices
        self.hough_rho = hough_rho
        self.hough_theta = hough_theta
        self.hough_thresh = hough_thresh
        self.min_line_len = min_line_len
        self.max_line_gap = max_line_gap
        self.filter_and_merge = filter_and_merge
        self.min_slope_thresh = min_slope_thresh
        self.slope_similarity_thresh = slope_similarity_thresh
        self.intercept_similarity_thresh = intercept_similarity_thresh
        self.extrapolate_lines = extrapolate_lines
        self.lane_horizon = lane_horizon
        self.side_margin = side_margin
        self.line_color = line_color
        self.line_thickness = line_thickness
        
    def grayscale(self, img):
        """Applies the Grayscale transform
        This will return an image with only one color channel
        but NOTE: to see the returned image as grayscale
        (assuming your grayscaled image is called 'gray')
        you should call plt.imshow(gray, cmap='gray')"""
        return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        # Or use BGR2GRAY if you read an image with cv2.imread()
        # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    def canny(self, img):
        """Applies the Canny transform"""
        return cv2.Canny(img, self.canny_thresh_lwr, self.canny_thresh_upr)

    def gaussian_blur(self, img):
        """Applies a Gaussian Noise kernel"""
        return cv2.GaussianBlur(img, (self.blur_kernel_size, self.blur_kernel_size), 0)

    def region_of_interest(self, img):
        """
        Applies an image mask.

        Only keeps the region of the image defined by the polygon
        formed from `vertices`. The rest of the image is set to black.
        `vertices` should be a numpy array of integer points.
        """
        #defining a blank mask to start with
        mask = np.zeros_like(img)   

        #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
        if len(img.shape) > 2:
            channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
            ignore_mask_color = (255,) * channel_count
        else:
            ignore_mask_color = 255

        #filling pixels inside the polygon defined by "vertices" with the fill color    
        cv2.fillPoly(mask, self.roi_vertices, ignore_mask_color)

        #returning the image only where mask pixels are nonzero
        masked_image = cv2.bitwise_and(img, mask)
        return masked_image

    def get_slope_intercept(self, lines):
        """
        Takes in a list of lines and returns corresponding slopes and intercepts for each
        """
        slopes = []
        intercepts = []
        for line in lines:
            for x1, y1, x2, y2 in line:
                slope = float(y2 - y1) / float(x2 - x1)
                intercept = float(y2) - float(slope * x2)
                slopes.append(slope)
                intercepts.append(intercept)

        return np.array(slopes), np.array(intercepts)

    def draw_lines(self, img, lines):
        """
        This function draws `lines` with `color` and `thickness`.    
        Lines are drawn on the image inplace (mutates the image).
        If you want to make the lines semi-transparent, think about combining
        this function with the weighted_img() function below
        """
        for line in lines:
            #print('line shape:', line.shape)
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), self.line_color, self.line_thickness)

    def filter_and_merge_lines(self, lines):
        """
        Filters out horizontal lines according to a minimum absolute slope and merges the 
        remaining lines by averaging groups formed through slope and intercept similarity
        """
        # Get slopes and intercepts
        slopes, intercepts = self.get_slope_intercept(lines)
        #print('slopes:', slopes)
        # Filter out horizontal lines
        filtered_indices = np.where(np.abs(slopes) >= self.min_slope_thresh)
        lines = np.array(lines)[filtered_indices]#.copy()
        slopes = slopes[filtered_indices]
        intercepts = intercepts[filtered_indices]

        # Establish an empty list to store indices of lines we have already merged
        deleted_indices = []
        # Establish an array to stack merged lines onto
        merged_lines = np.zeros((0, 1, 4), dtype='int32')

        # Iterate through remaining arrays and either average groups or add unique lines
        for i, (slope, intercept, line) in enumerate(zip(slopes, intercepts, lines)):
            #print('len slopes:', len(slopes))
            # Check to see if index has already been assigned to a group and skip if so
            if i in deleted_indices:
                continue
            # Break if we've merged all lines
            if len(slopes) <= 0:
                break
            #print('new slopes:', slopes)
            #print('current slope:', slope)
            #print('abs(new slopes - current slope):', np.abs(slopes-slope))
            # Get indices of lines similar in slope to current index
            slope_group_indices = np.where(np.abs(slopes - slope) 
                                           < self.slope_similarity_thresh)[0]
            #print('slope_group_indices:', slope_group_indices)
            # Get indices of lines with similar intercepts
            #print('abs(intercepts - current intercept):', np.abs(intercepts - intercept))
            int_group_indices = np.where(np.abs(intercepts - intercept) 
                                         < self.intercept_similarity_thresh)[0]
            # Find intersection of these groups
            group_indices = [idx for idx in slope_group_indices if idx in int_group_indices]
            #print('group_indices:', group_indices)
            # If there are no similar lines, stack current unique line and move on
            if len(group_indices) == 0:
                group_indices = [0]
                #merged_lines = np.vstack([merged_lines, line.reshape((1, 1, 4))])
                #slopes = np.delete(slopes, 0, 0)
                #intercepts = np.delete(intercepts, 0, 0)
                #lines = np.delete(lines, 0, 0)
                #deleted_indices.append(i)
                #continue
            # Otherwise average the lines in the group, stack merged line,
            # and update deleted indices accordingly
            merged_line = lines[group_indices].mean(axis=0)
            #print('merged_line:', merged_line)
            #print('merged_line shape:', merged_line.shape)
            merged_line = merged_line.round(0).astype('int32')
            #print('merged_line after round:', merged_line)
            slopes = np.delete(slopes, group_indices, 0)
            intercepts = np.delete(intercepts, group_indices, 0)
            lines = np.delete(lines, group_indices, 0)
            [deleted_indices.append(idx+i) for idx in group_indices]
            merged_lines = np.vstack([merged_lines, merged_line.reshape((1, 1, 4))])

        #print('merged_lines:', merged_lines)
        return merged_lines

    def hough_lines(self, img, line_img=None):
        """
        `img` should be the output of a Canny transform.

        Optional params:
        filter_and_merge - (bool) whether to filter and merge lines before drawing
        min_slope - (float) minimum slope of lines to be drawn
        slope_similarity_thresh - (float) max distance between slope values to merge lines
        intercept_similarity_thresh - (int) max number of pixels between intercepts to merge lines

        Returns an image with hough lines drawn.
        """
        #plt.imshow(img)
        lines = cv2.HoughLinesP(img, 
                                self.hough_rho, 
                                self.hough_theta, 
                                self.hough_thresh, 
                                np.array([]), 
                                minLineLength=self.min_line_len, 
                                maxLineGap=self.max_line_gap,
                               )
        #print('lines:', lines)
        
        #if line_img is None:
        #    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
        #self.draw_lines(line_img, lines)
        
        return lines

    # Python 3 has support for cool math symbols.

    def weighted_img(self, img, initial_img, =0.8, =1., =0.):
        """
        `img` is the output of the hough_lines(), An image with lines drawn on it.
        Should be a blank image (all black) with lines drawn on it.

        `initial_img` should be the image before any processing.

        The result image is computed as follows:

        initial_img *  + img *  + 
        NOTE: initial_img and img must be the same shape!
        """
        return cv2.addWeighted(initial_img, , img, , )
    
    def extrapolate(self, lines, imshape):
        slopes, intercepts = self.get_slope_intercept(lines)
        new_lines = []
        for slope, intercept in zip(slopes, intercepts):
            y1 = imshape[0]
            x1 = int(round((y1 - intercept) / slope, 0))
            y2 = imshape[0] * self.lane_horizon
            x2 = int(round((y2 - intercept) / slope, 0))
            new_lines.append([[x1, y1, x2, int(y2)]])
            
        return np.array(new_lines)
    
    def process_image(self, initial_img):
        # Generate default roi vertices if none provided
        imshape = initial_img.shape
        if self.roi_vertices is None:
            self.roi_vertices = np.array([[(int(imshape[1] * self.side_margin), 
                                            imshape[0]),
                                           (imshape[1] // 2 - imshape[1] // 20, 
                                            #imshape[0] // 2 + imshape[0] // 10),
                                            int(imshape[0] * self.lane_horizon)), 
                                           (imshape[1] // 2 + imshape[1] // 20, 
                                            #imshape[0] // 2 + imshape[0] // 10),
                                            int(imshape[0] * self.lane_horizon)),
                                           (imshape[1] - int(imshape[1] * self.side_margin), 
                                            imshape[0])]], 
                                         dtype=np.int32)
            #print(self.roi_vertices)
        # Make grayscale copy of input image
        gray = self.grayscale(initial_img.copy())
        # Blur image
        blur_gray = self.gaussian_blur(gray)
        # Get Canny edge detection
        edges = self.canny(blur_gray)
        #plt.imshow(edges)
        #plt.show()
        # Make mask and get RoI
        masked_edges = self.region_of_interest(edges)
        #plt.imshow(masked_edges)
        #plt.show()
        # Run masked edges through Hough transform to get lines           
        lines = self.hough_lines(masked_edges)
        # Filter lines if appropriate
        if self.filter_and_merge:
            lines = self.filter_and_merge_lines(lines)
        # Extrapolate lines if appropriate
        if self.extrapolate_lines:
            lines = self.extrapolate(lines, imshape)
            
        # Draw lines on image and return
        if self.filter_and_merge:
            line_img = initial_img
            self.line_color = [255,0,255]
            self.line_thickness = 4
        else:
            line_img = np.zeros((imshape[0], imshape[1], 3), dtype=np.uint8)
            self.line_color = [255,0,0]
            self.line_thickness = 2
        self.draw_lines(line_img, lines)
        #return line_img
        if self.filter_and_merge:
            return line_img
        else:
            return self.weighted_img(line_img, initial_img)
````
